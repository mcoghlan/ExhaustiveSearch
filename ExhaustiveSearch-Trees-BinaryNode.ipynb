```python
def find_value(node="", target=""):
     """that takes a node and target value as parameters.
     Method searchs the nodeâ€™s subtree for the target and prints a string 
     indicating whether it found the value."""
     result = node.find_node(target)
     if result == "found": print("Found " + target)
     else: print("Value " + target + " not found")



class BinaryNode:

    indent = "  "
    result = ""

    def find_node(self, target=""): 
        """ exhaustively search the tree for a target value """

        working_target = target
        result = ""
        try_target=""
        # print("\nNode is: " + self.value + " and target to be found is: " + working_target)
       
        if (self != None) or (self.left_child != None) or (self.right_child != None): 

            if (self.value == working_target): 
                #or (self.left_child.value == working_target) or (self.right_child.value == working_target): 
                #print("Node value is: "+ self.value + " and left_child.value = " + self.left_child.value)
                #print("Target is still: " + working_target)
                result = "found"
                return result
                                         

            if self.left_child != None:
                result = self.left_child.find_node(working_target)
                if result == "found": 
                    return result
        
            if self.right_child != None:
                result = self.right_child.find_node(working_target) 
                if result == "found":
                    return result
            
            result = "Value " + working_target + " not found"
    
        return result



# result += self.left_child.__str__(level + 1)
    def add_left(self, left):
        self.left_child = left

    def add_right(self, right):
        self.right_child = right

    def __str__(self, level=0):
        
        result = ""

        if level == 0: 
            result = "\n" + self.value + ":"
            level += 1
  

        if (self.left_child != None) or (self.right_child != None): 

            # left child 
            if (self.left_child == None) and (self.right_child != None):
                result += "\n" + level*self.indent + "None"
            else:
                result += "\n" + level*self.indent + self.left_child.value +":"
                result += self.left_child.__str__(level+1)
                

            # right child                         
            if (self.right_child == None) and (self.left_child != None):
                result += "\n" + level*self.indent + "None"
            else:                
                result += "\n" + level*self.indent + self.right_child.value + ":" 
                result += self.right_child.__str__(level+1)
                
        print(result)
        return result

    
    def __init__(self, value):
        self.value = value
        self.left_child = None
        self.right_child = None
        return

```

```python
root = BinaryNode('Root')
a = BinaryNode('A')
b = BinaryNode('B')
c = BinaryNode('C')
d = BinaryNode('D')
e = BinaryNode('E')
f = BinaryNode('F')

root.add_left(a)
root.add_right(b)
a.add_left(c)
a.add_right(d)
b.add_right(e)
e.add_left(f)

```
